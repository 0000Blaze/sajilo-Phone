```python
import os
import pandas as pd

csv_folder = '/media/rohan/DATA/project/saziloPhonebook/data/processed_data'
csv_files = [f for f in os.listdir(csv_folder) if f.endswith('.csv')]

df_dict = {}

for file in csv_files:
    filepath = os.path.join(csv_folder, file)
    df = pd.read_csv(filepath, skiprows=[1])  # skip the first row
    filename = os.path.splitext(file)[0]
    df_dict[filename] = df

print(df_dict)
```
