```python
data = data.iloc[1:169,0:12]
data.info()
```

```python
# data.iloc[2]
columnNames = ['District','Mayor','MayorNum','DeputyMayor','DeputyMayorNum','ChiefExecutiveOfficer','ChiefExecutiveOfficerNum','InformationOfficer','InformationOfficerNum','OfficeContact','Delete','NepaliDistrictName']
data.columns = columnNames
data = data.drop('Delete', axis=1)
```

```python
data = data.dropna(how='all') # remove all NaN row entries
data.info()
```

```python
#  Have required column name and row entries , now need to break them down
data.head(15)
```

```python
# Break the dataframe into district parts
# initialize list to store indices
indices = []
# iterate over rows
for i, row in data.iterrows():
    # check if first column is not NaN and all other columns are NaN
    if pd.notna(row['District']) and pd.isna(row.iloc[1:]).all():
        # if condition is True, add the row index to the list of indices
        indices.append(i)
```

```python
# correction of indices, error due to removal of blank rows and unnecessary rows
indices=[0,10,19,30,46,64,77,85,92,103,113,122,133,142]
indices
```

```python
# split dataframe into smaller dataframes based on district
dfs = []
# dfs.append(data.iloc[0:indices[i], :])
for i in range(len(indices)-1):
    dfs.append(data.iloc[indices[i]:indices[i+1], :])

dfs.append(data.iloc[indices[-1]:, :])
```

```python
#save each district as csv
for df in dfs:
    filename = 'processed_data/'+ str(df.iloc[0,0]) + '.csv'
    df.to_csv(filename, index=False)
```
